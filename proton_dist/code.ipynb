```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("bmh")
```

```python
dist = pd.read_csv("proton_dist2.csv", header=None, names=["energy", "flux"])
```

```python
plt.scatter(dist["energy"], dist["flux"])
plt.xscale('log'); plt.yscale('log');
```

```python
x = np.logspace(
    np.log10(dist["energy"].min()), 
    np.log10(dist["energy"].max())
)
x
```

```python
pfit = np.poly1d(np.polyfit(dist["energy"], dist["flux"], 10))
print(pfit)

plt.plot(x, pfit(x))
plt.scatter(dist["energy"], dist["flux"])
plt.xscale('log'); plt.yscale('log');
```
